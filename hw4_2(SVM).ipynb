```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC # support vector classification
```

```python
# iris data set
iris = datasets.load_iris()
```

```python
# Divide the data and target in the Iris datasets by the variables X,y.

X = iris.data   # sepal length (cm)  sepal width (cm)  petal length (cm)  petal width (cm)
y = iris.target # setosa=0, versicolor=1, virginica=2
```

```python
# Set the dataframe
dfx = pd.DataFrame(X, columns = iris.feature_names)

x_pl = dfx['petal length (cm)'] # 꽃잎의 길이
x_pw = dfx['petal width (cm)'] # 꽃잎의 너비
```

```python
svm = SVC(kernel = 'rbf', gamma = 'auto')
svm.fit(X,y)
```

```python
# iris data와 target간의 평균 정확도
svm.score(X,y)
```

```python
plt.scatter(x_pl,x_pw,c=svm.predict(X))
plt.xlabel('petal length (cm)') 
plt.ylabel('petal width (cm)')
plt.show()
```

보라색이 setosa, 청록색이 versicolor, 노란색이 virginica 이다.
표의 x축은 꽃잎의 길이, y축은 꽃잎의 너비이다.
위의 두 변수로 setosa는 versicolor와 virginica 끼리는 확실하게 차이가 나타남을 볼 수 있지만, versicolor와 virginica는 서로 중첩되는 구간이 생긴다. 
데이터가 중첩되는 구간을 제외한 나머지 값끼리는 versicolor와 virginica도 어느정도 구분해낼 수 있다.
